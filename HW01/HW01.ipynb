```python
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

## 1a

```python
# mean and std
mx, sigmax = 100, 0.2
my, sigmay = 230, 0.3
mz, sigmaz = 135, 0.1

def get_random_values(mu, sigma, n):
    # get random coordinates
    random.seed(42)
    return [random.gauss(mu, sigma) for _ in range(n)]

x_values = get_random_values(mx, sigmax, n = 300)
y_values = get_random_values(my, sigmay, n = 300)
z_values = get_random_values(mz, sigmaz, n = 300)
```

```python
"""1a"""
# plot 300 points in 3D figure
fig = plt.figure(figsize = (8,8)) 
ax = plt.axes(projection ="3d")
plt.ticklabel_format(useOffset=False)
plt.tight_layout()
ax.scatter3D(x_values, y_values, z_values, color = "green")
ax.set_xlabel('X-axis', fontweight ='bold', fontsize=20, labelpad=20) 
ax.set_ylabel('Y-axis', fontweight ='bold', fontsize=20, labelpad=30) 
ax.set_zlabel('Z-axis', fontweight ='bold', fontsize=20, labelpad=10)
ax.set_title("300 random points", fontweight ='bold', fontsize=20, loc="center", y = 1.05)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.tick_params(axis='z', labelsize=12)
# ax.view_init(elev=3, azim=10)
# plt.savefig('1a.jpg', bbox_inches='tight')
```

## 1b

```python
"""1b"""
print(f"x_std:{round(np.std(x_values, ddof=1),5)}")
print(f"y_std:{round(np.std(y_values, ddof=1),5)}")
print(f"z_std:{round(np.std(z_values, ddof=1),5)}")
print(f"x_mean:{round(np.mean(x_values),5)}")
print(f"y_mean:{round(np.mean(y_values),5)}")
print(f"z_mean:{round(np.mean(z_values),5)}")
```

```python
"""1c"""
fig = plt.figure(figsize = (8,8)) 
ax = plt.axes(projection ="3d")
plt.ticklabel_format(useOffset=False)
plt.tight_layout()
ax.scatter3D(x_values, y_values, z_values, color = "green")
ax.set_xlabel('X-axis', fontweight ='bold', fontsize=20, labelpad=20) 
ax.set_ylabel('Y-axis', fontweight ='bold', fontsize=20, labelpad=30) 
ax.set_zlabel('Z-axis', fontweight ='bold', fontsize=20, labelpad=10)
ax.set_title("300 random points with 1 (red color)", fontweight ='bold', fontsize=20, loc="center", y = 1.05)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.tick_params(axis='z', labelsize=12)

x_max, x_min = mx+sigmax, mx-sigmax
y_max, y_min = my+sigmay, my-sigmay
z_max, z_min = mz+sigmaz, mz-sigmaz
x = [x_min, x_max, x_max, x_min, x_min, x_max, x_max, x_min]
y = [y_min, y_min, y_max, y_max, y_min, y_min, y_max, y_max]
z = [z_min, z_min, z_min, z_min, z_max, z_max, z_max, z_max]
pts = list(zip(x,y,z))
verts = [[pts[0],pts[1],pts[2],pts[3]],
         [pts[4],pts[5],pts[6],pts[7]],
         [pts[0],pts[1],pts[5],pts[4]],
         [pts[2],pts[3],pts[7],pts[6]],
         [pts[1],pts[2],pts[6],pts[5]],
         [pts[4],pts[7],pts[3],pts[0]]]
ax.add_collection3d(Poly3DCollection(verts, facecolors='red', alpha=.20))
# plt.show()
plt.savefig('1c.jpg', bbox_inches='tight')
```

## 1d

```python
"""1d"""
def draw_1d(n):
    x_values = get_random_values(mx, sigmax, n = n)
    y_values = get_random_values(my, sigmay, n = n)
    z_values = get_random_values(mz, sigmaz, n = n)
    print(f"n = {n}")
    print(f"x_std:{np.std(x_values, ddof=1)}")
    print(f"y_std:{np.std(y_values, ddof=1)}")
    print(f"z_std:{np.std(z_values, ddof=1)}")
    print(f"x_mean:{np.mean(x_values)}")
    print(f"y_mean:{np.mean(y_values)}")
    print(f"z_mean:{np.mean(z_values)}")

    fig = plt.figure(figsize = (8,8)) 
    ax = plt.axes(projection ="3d")
    plt.ticklabel_format(useOffset=False)
    plt.tight_layout()
    ax.scatter3D(x_values, y_values, z_values, color = "green")
    ax.set_xlabel('X-axis', fontweight ='bold', fontsize=20, labelpad=20) 
    ax.set_ylabel('Y-axis', fontweight ='bold', fontsize=20, labelpad=30) 
    ax.set_zlabel('Z-axis', fontweight ='bold', fontsize=20, labelpad=10)
    ax.set_title(f"{n} random points", fontweight ='bold', fontsize=20, loc="center", y = 1.05)
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    ax.tick_params(axis='z', labelsize=12)
    plt.savefig(f'1d_{n}.jpg', bbox_inches='tight')

for n in [3000,30000,300000,3000000]:
    draw_1d(n)
```

## 3

```python
"""3"""
import networkx as nx
import matplotlib.pylab as pl

# directed graph
plt.figure(figsize=(8,8))
G=nx.DiGraph()

# add 10 points
for i in range(1,11):
    G.add_node(i)

# config the network
edges = [[i, i+1] for i in range(1,10)] + [[6,1]] + [[3, 10]]
for edge in edges:
    G.add_edge(*edge)

# relabel the points as P1, P2...
G = nx.relabel_nodes(G, lambda node: f"P{node}", copy=False)

# get positions
pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=400, font_size=12,with_labels=True)

# shift position a little bit
shift = [0.05, 0]
shifted_pos ={node: node_pos + shift for node, node_pos in pos.items()}

# add elev info to print in addition to node ids
labels = {}
random.seed(42)
h1 = round(random.uniform(2,8), 3)
h2 = round(random.uniform(2,8), 3)
h3 = round(random.uniform(2,8), 3)
h4 = round(random.uniform(2,8), 3)
h5 = round(random.uniform(2,8), 3)
h6 = round(random.uniform(2,8), 3)
h7 = round(random.uniform(2,8), 3)
h8 = round(random.uniform(2,8), 3)
h9 = round(random.uniform(2,8), 3)
h10 = round(random.uniform(2,8), 3)
labels["P1"] = f"h1={h1}m"
labels["P2"] = f"h2={h2}m"
labels["P3"] = f"h3={h3}m"
labels["P4"] = f"h4={h4}m"
labels["P5"] = f"h5={h5}m"
labels["P6"] = f"h6={h6}m"
labels["P7"] = f"h7={h7}m"
labels["P8"] = f"h8={h8}m"
labels["P9"] = f"h9={h9}m"
labels["P10"] = f"h10={h10}m"

nx.draw_networkx_labels(G, shifted_pos, labels=labels, horizontalalignment="left")


# adjust frame to avoid cutting text, may need to adjust the value
axis = pl.gca()
axis.set_xlim([1.2*x for x in axis.get_xlim()])
# axis.set_ylim([1.5*y for y in axis.get_ylim()])
# plt.savefig('graph.jpg', dpi=200)
```

```python
def get_elev_diff(elevations, edges, i):
    # compute true elev diff
    return round(elevations[edges[i][1]-1] - elevations[edges[i][0]-1], 3)


def add_rand_error(true_elev_diff):
    #add 7mm/km as random error
    random.seed(42)
    error_elev_diff = [round(h + random.uniform(-7 / 1000, 7 / 1000), 3) \
        for h in true_elev_diff]

    return error_elev_diff

elevations = [h1,h2,h3,h4,h5,h6,h7,h8,h9,h10]
true_elev_diffs = [get_elev_diff(elevations, edges, i) for i in range(len(edges))]
error_elev_diffs = add_rand_error(true_elev_diffs)
true_errors = [round(error - true, 3) for error, true in zip(error_elev_diffs ,true_elev_diffs)]
print(edges)
print(true_elev_diffs)
print(error_elev_diffs)
print(true_errors)
```

```python
import csv
with open('output.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['BS_Pt', 'FS_Pt', 'elevation difference(True)', 'elevation difference (with error)', 'true error'])
    for edge, true_elev_diff, error_elev_diff, true_error in \
        zip(edges, true_elev_diffs, error_elev_diffs, true_errors):
        bs_pt, fs_pt = list(map(lambda x: "P"+str(x), edge))
        writer.writerow([bs_pt, fs_pt, true_elev_diff, error_elev_diff, true_error])
```
